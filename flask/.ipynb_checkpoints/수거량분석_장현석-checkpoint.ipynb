```python
import numpy as np
import pandas as pd
from plotnine import *
import matplotlib.pyplot as plt
# jupyter notebook warning 무시하기

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['axes.unicode_minus'] = False

from keras.utils.vis_utils import model_to_dot
from keras.models import Sequential
from keras.layers import Dense, Activation, LSTM, Dropout
import statsmodels.api as sm
from scipy import stats
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import mglearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
from IPython.display import Image
import graphviz
import pydotplus
```

```python
#파일 불러오기
df1 = pd.read_csv('../team/수질+수거량+정화_r.csv',engine='python')
```

```python
df1['해수수질기준'] = df1['해수수질기준'].astype('float')
```

```python
# 결측치와 사용하지않을 행 제거
df1 = df1.drop('구군',axis=1)
df1 = df1.drop('Ds',axis=1)
df1 = df1.drop('정화면적',axis=1)
```

## keras

```python
# 쓰레기수거량 = 수거량 + 수거량2
x = df1.iloc[:,36:38] 
y = df1.iloc[:,4]
Y = pd.get_dummies(y).values
X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size=0.1, random_state=1)
```

```python
# 모델생성
model = Sequential()
model.add(Dense(500, input_shape=(2,), activation='relu'))
model.add(Dropout(0.15))
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.15))
model.add(Dense(5, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# 모델학습
hist1 = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
```

```python
model.save('model_waste.h5')
```

```python
# 1. 훈련세트의 각 특성의 평균계산
mean_on_train = X_train.mean(axis=0)

# 2. 훈련세트의 각 특성의 표준편차계산
std_on_train = X_train.std(axis=0)
# print(mean_on_train)
# print(std_on_train)

# 3. 데이터에서 평균을 뺴고 표준편차로 나누면
#    평균0, 표준편차가 1인 데이터로 변환이 된다.
X_train_scaled = (X_train - mean_on_train) / std_on_train
# print(X_train_scaled)

# 4. 훈련용데이터셋에도 동일 계산을 적용
mean_on_test = X_test.mean(axis=0)
std_on_test = X_test.std(axis=0)
X_test_scaled = (X_test - mean_on_test) / std_on_test
# X_test_scaled = (X_test - mean_on_train) / std_on_train

# 5. 기계학습
mlp = MLPClassifier(random_state=30).fit(X_train_scaled, y_train)

# 6. 결과출력

X_test_scaled = (X_test - mean_on_train) / std_on_train
```

```python
# 최적화알고리즘을 변경 (adam -> lbfgs)
mlp = MLPClassifier(solver="lbfgs", random_state=30).fit(X_train_scaled, y_train)
```

```python
# 특성중요도 시각화
def plot_feature_importances_cancer(model):
    n_features = x.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align="center")
    plt.yticks(np.arange(n_features),x)
    plt.xlabel("특성중요도")
    plt.ylabel("특성")
    plt.ylim(-1, n_features)
    plt.show()
    
plot_feature_importances_cancer(forest)


```

## 회귀분석

```python
# 귀무가설 : 해양쓰레기수거량에 따라 수질등급에 영향이 없다.
# 대립가설 : 해양쓰레기수거량에 다라 수질등급에 영향이 있다.
x = df1.iloc[:,36:38]
y = df1.iloc[:,4]

model = LinearRegression().fit(x, y) 
predicted = model.predict(x) 

plt.scatter(y, predicted)
plt.xlabel("수질성분")
plt.ylabel("수질등급")
plt.title("수질성분과 수질등급")

result = stats.linregress(y, predicted)

slope, intercept, r_value, p_value, stderr = result
x1 = np.array(y)
plt.plot(x1, slope*x1+intercept, c="red")
plt.show()

# 결과분석
# rvalue = 0.31 / 31%의 설명력이 있다.
# 대립가설을 기각하고 귀무가설을 채택한다.
# 즉, 쓰레기수거량에 따라 수질등급에 영향이 없다.
```
