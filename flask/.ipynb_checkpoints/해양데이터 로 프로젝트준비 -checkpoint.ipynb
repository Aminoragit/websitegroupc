```python
df = pd.read_csv('../data/haeyang/수질.수거량.csv')
df = df.drop(['Ds'], axis=1)

data=df
```

```python
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    # noise term prevents the zero division
    return numerator / (denominator + 1e-7)
```

```python
X = data.iloc[:,5:35]
y = data.iloc[:,4]
Y = pd.get_dummies(y).values

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 모델생성
model = Sequential()
model.add(Dense(64, input_shape=(30,), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(5, activation='softmax'))
model.compile('adam', 'categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# 모델학습
hist = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
```

```python
# 시각화
plt.figure(figsize=(12,8))
plt.plot(hist.history['loss']) # 학습용 데이더셋의 손실률
plt.plot(hist.history['val_loss']) # 검증용 데이더셋의 손실률
plt.plot(hist.history['accuracy']) # 학습용 데이더셋의 정확도
plt.plot(hist.history['val_accuracy']) # 학습용 데이더셋의 정확도

plt.legend(['loas', 'val_loss', 'acc', 'val_acc'])

plt.show()
```

```python
# 모델평가
loss, accuracy = model.evaluate(X_test, y_test)
print('정확도={:.2f}'.format(accuracy))
```

```python
# 모델예측 : 오분류보고서
y_pred = model.predict(X_test)
y_pred
y_test_class = np.argmax(y_test, axis=1)
print(y_test_class)
y_pred_class = np.argmax(y_pred, axis=1)
print(y_pred_class)

print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```

```python
caler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)

# 그림객체 생성
fig, axes = plt.subplots(1,4, figsize=(15,3), subplot_kw={'xticks':(), 'yticks':()})

# 사용할 머신러닝알고리즘 모델을 리스트로 생성
algorithms = [KMeans(n_clusters=2), AgglomerativeClustering(n_clusters=2), DBSCAN()]

# 각 알고리즘의 비교를 위해 무작위로 클러스터 할당을 한다.
random_state = np.random.RandomState(seed=0)
random_clusters = random_state.randint(low=0, high=2, size=len(X))

# 무작위로 할당한 클러스터를 시각화
axes[0].scatter(X_scaled[:,0], X_scaled[:,1], c=random_clusters, cmap=mglearn.cm3, s=60, edgecolor='k')
axes[0].set_title("무작위할당-ARI:{:.2f}".format(adjusted_rand_score(y, random_clusters)))

for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    ax.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters, cmap=mglearn.cm3, s=60, edgecolor='k')
    ax.set_title("{}-ARI:{:.2f}".format(algorithm.__class__.__name__,
                                        adjusted_rand_score(y, clusters)))
plt.show()

# 결과분석
# 클러스터를 무작위로 할당했을 때의 ARI점수는 0이고, DBSCAN은 점수가 1(완변한 군집)이다.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("훈련세트 점수: {:.2f}".format(lr.score(X_train, y_train)))
print("검증세트 점수: {:.2f}".format(lr.score(X_test, y_test)))
# 결과분석
# 훈련세트와 검증세트의 점수를 비교해 보면 훈련세트의 예측은 매우 정확한 반면에
# 검증세트는 R2값이 더높다(60%). 이런 성능차이는 모델이 과대적합이 되었다는 
# 의미이기 때무에 이런 복잡한 특성치를 제어할 수 있는 모델을 사용해야 하는데 
# 기본 선형회귀방식대신 가장 널리 사용되는 모델은 릿지회귀이다.
```
